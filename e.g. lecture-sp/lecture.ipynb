### セリーグ防御率

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

# 2021年の防御率データ
era_data_2021 = {
    'ヤクルト': 3.81,
    '阪神': 3.30,
    '巨人': 3.65,
    '広島': 3.82,
    '中日': 3.22,
    'DeNA': 3.76
}

# チーム名と防御率をリストに格納
teams = list(era_data_2021.keys())
eras = list(era_data_2021.values())

# 棒グラフを作成
plt.figure(figsize=(10, 6))
plt.bar(teams, eras, color=['deepskyblue', 'gold', 'orange', 'crimson', 'navy', 'royalblue'])

# タイトルやラベルを設定
plt.title('2021年 セ・リーグ 各球団の防御率', fontsize=16)
plt.xlabel('球団名', fontsize=12)
plt.ylabel('防御率', fontsize=12)
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# グラフを表示
plt.show()
```
